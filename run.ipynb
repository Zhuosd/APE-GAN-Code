```
import torch
```

```
fgsm1 = torch.load('./MNIST_AE/Training_Set/fgsm1.tar')
x1 , t1, adversarial1 = fgsm1['normal_data_x'], fgsm1['normal_data_t'], fgsm1['adv']

fgsm3 = torch.load('./MNIST_AE/Training_Set/fgsm3.tar')
x3 , t3, adversarial3 = fgsm3['normal_data_x'], fgsm3['normal_data_t'], fgsm3['adv']

fgsm5 = torch.load('./MNIST_AE/Training_Set/fgsm5.tar')
x5 , t5, adversarial5 = fgsm5['normal_data_x'], fgsm5['normal_data_t'], fgsm5['adv']

fgsm7 = torch.load('./MNIST_AE/Training_Set/fgsm7.tar')
x7 , t7, adversarial7 = fgsm7['normal_data_x'], fgsm7['normal_data_t'], fgsm7['adv']

x = torch.cat((x1,x3,x5,x7))
t = torch.cat((t1,t3,t5,t7))
adversarial = torch.cat((adversarial1,adversarial3,adversarial5,adversarial7))
torch.save({"normal_data_x": x, "normal_data_t":t, "adv": adversarial}, "data.tar")
```
